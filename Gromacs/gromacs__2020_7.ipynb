## **Gromacs-2020 download and install**

```python
#@title ### **Checking CPU and GPU**
#@markdown Run to Check CPU & GPU
%%bash
lscpu
nvidia-smi
```

```python
#@title ### **Download and Extract CMAKE**
#@markdown Run to Download and Extract CMAKE
%%bash
wget -c -N https://github.com/Kitware/CMake/releases/download/v3.24.3/cmake-3.24.3.tar.gz
tar -zxvf cmake-3.24.3.tar.gz
```

```python
#@title ### **Install CMAKE**
#@markdown Run to install CMAKE. Will take some time...........
%%bash
mkdir /content/cmake-3.24.3/build
cd /content/cmake-3.24.3/build
cmake /content/cmake-3.24.3
make -j 4   ## use 4 cpu
make install
```

```python
#@title ### **Download and Extract Gromacs-2020.7**
#@markdown Run to Download and Extract Gromacs 2020.7
%%bash
wget -c -N https://ftp.gromacs.org/gromacs/gromacs-2020.7.tar.gz
tar -zxvf gromacs-2020.7.tar.gz
```

```python
#@title ### **Install Gromacs-2020.7**
#@markdown Need to Wait. Installing...

%%bash
mkdir /content/gromacs-2020.7/build
cd /content/gromacs-2020.7/build
cmake .. -DGMX_BUILD_OWN_FFTW=ON -DREGRESSIONTEST_DOWNLOAD=ON -DGMX_GPU=CUDA
make -j 4
make install
```

```python
#@title ### **Checking Gromacs is Installed**
#@markdown Run to check
%%bash
source /usr/local/gromacs/bin/GMXRC

gmx -h
```
