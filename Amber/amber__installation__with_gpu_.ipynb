```python
# Install Amber
%%bash

cd /content/amber22_src/build
# ./run_cmake
cmake /content/amber22_src \
    -DCMAKE_INSTALL_PREFIX=/content/amber22 \
    -DCOMPILER=GNU  \
    -DMPI=TRUE -DCUDA=TRUE -DINSTALL_TESTS=TRUE \
    -DDOWNLOAD_MINICONDA=TRUE \
    2>&1 | tee  cmake.log
    
make -j 2  # Number of CPU
make install
   
```
